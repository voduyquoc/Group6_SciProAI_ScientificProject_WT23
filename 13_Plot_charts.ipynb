```python
annual_returns_df = pd.read_csv("annual_returns_df.csv", index_col=[0])
annual_volatility_df = pd.read_csv("annual_volatility_df.csv", index_col=[0])
```

```python
annual_returns_df.index = annual_returns_df.index.astype(str)
annual_volatility_df.index = annual_volatility_df.index.astype(str)
```

```python
# Create a figure and axes for subplots
fig, axs = plt.subplots(1, 5, figsize=(20, 5))

# Initialize variables to store y-axis limits
y_min = float(0.0)
y_max = float(0.25)

# Iterate over the columns
for i, (returns_col, volatility_col) in enumerate(zip(annual_returns_df.columns, annual_volatility_df.columns)):
    # Plot combined chart for each model
    ax = axs[i]
    # Set x-axis ticks to only show specific values
    ax.bar(annual_returns_df.index, annual_returns_df[returns_col], color='blue', label='Returns', width=0.7)
    ax.plot(annual_volatility_df.index, annual_volatility_df[volatility_col], color='red', marker='o', label='Volatility')
    ax.set_title(f'{returns_col}', fontsize=15)
    ax.legend(loc='lower left')
    ax.set_xlabel('Number of features', fontsize=13)
    ax.set_ylabel('Returns / Volatility', fontsize=13)
    
    # Update y-axis limits
    y_min = min(y_min, annual_returns_df[returns_col].min(), annual_volatility_df[volatility_col].min())
    y_max = max(y_max, annual_returns_df[returns_col].max(), annual_volatility_df[volatility_col].max())

# Set y-axis limits for all charts
for ax in axs:
    ax.set_ylim(y_min, y_max)

# Adjust layout
plt.tight_layout()

plt.savefig('34_compare.png',bbox_inches='tight')

# Show the plots
plt.show()
```
